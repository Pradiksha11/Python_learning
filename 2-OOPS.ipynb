```python
#OOP - Classes and Objects
class Student:
    def say_hello(self):#method , 
        #self means it tell the python who is the caller of this method (s1)
        print("Hi, I'm Pradiksha")

s1 = Student() #s1 object creates a memory 
s1.say_hello() #s1.say_hello(s1) like this it will call
```

```python
#with constructor
class Student:
    def __init__(self,name,age): #init is constructor
        self.name = name
        self.age = age
    def display (self):
        print(f"Name: {self.name}, Age: {self.age}")    

s1=Student("pradiksha","25")
s1.display()

#without constructor method argument call
class Student:
    def add (self,a,b):
        return a+b

s1=Student()
print(s1.add(2,3))

```

```python
#Abstraction 
from abc import ABC,abstractmethod

class Score_card(ABC):
    @abstractmethod
    def subjects(self):
        print("Parent content")
        return(['marks','average'])
        

    @abstractmethod
    def marks(self):
        pass

class result(Score_card):

    def subjects(self):
        p = super().subjects()
        print(p+["sum"])
    def marks(self):
        print("two method")

res=result()
res.subjects()

        
    

```
