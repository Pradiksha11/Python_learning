```python
#Abstraction 
from abc import ABC,abstractmethod

class Score_card(ABC):
    @abstractmethod
    def subjects(self):
        print("Parent content")
        return(['marks','average'])
        

    @abstractmethod
    def marks(self):
        pass

class result(Score_card):

    def subjects(self):
        p = super().subjects()
        print(p+["sum"])
    def marks(self):
        print("two method")

res=result()
res.subjects()

        
    

```

```python
#access specifier in variables
class parent:
    def __init__(self):
        self.pub_var = "Public"
        self._prot_var = "Protected"
        self.__priv_var = "Private"

    def acc_same_class(self):
        print(self.pub_var)
        print(self._prot_var)
        print(self.__priv_var)

class child_class(parent):
    def acc_sub_class(self):
        print(self.pub_var)
        print(self._prot_var)
        try:
           print(self.__priv_var)
        except AttributeError:
            print("Private variable cannot be used in the sub child class")   

#to work on the other files instead we created
# a normal class but while creating a class we passing the object of the parent class

class other_file_class:
    def acc_other_fil(self,obj):
        print(obj.pub_var)
        print(obj._prot_var)
        print(obj._parent__priv_var) #name mangling
        try:
            print(obj.__priv_var)
        except:
            print("Other calling")
        

a=parent()
a.acc_same_class()
b=child_class()
b.acc_sub_class()
print(dir(a))
c=other_file_class()
c.acc_other_fil(a)
```

```python
#access specifier in methods
class parent:
    def pub_meth(self):
        print("Public methos")

    def __prv_meth(self):
        print("Private methos")

    def _pro_meth(self):
        print("Protected methos")

    def acc_same_class(self):
        print(self.pub_meth())
        print(self.__prv_meth())
        print(self._pro_meth())

class child_class(parent):
    def acc_sub_class(self):
        print(self.pub_meth())
        print(self._pro_meth())
        try:
           print(self.__prv_meth())
        except AttributeError:
            print("Private method cannot be used in the sub child class")   

#to work on the other files instead we created
# a normal class but while creating a class we passing the object of the parent class

class other_file_class:
    def acc_other_fil(self,obj):
        print(obj.pub_meth())
        print(obj._pro_meth())
        print(obj._parent__prv_meth()) #name mangling
        try:
            print(obj.__prv_meth())
        except:
            print("Other calling")
        

a=parent()
a.acc_same_class()
b=child_class()
b.acc_sub_class()
c=other_file_class()
c.acc_other_fil(a)
```
