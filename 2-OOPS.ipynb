```python
#Abstraction 
from abc import ABC,abstractmethod

class Score_card(ABC):
    @abstractmethod
    def subjects(self):
        print("Parent content")
        return(['marks','average'])
        

    @abstractmethod
    def marks(self):
        pass

class result(Score_card):

    def subjects(self):
        p = super().subjects()
        print(p+["sum"])
    def marks(self):
        print("two method")

res=result()
res.subjects()

        
    

```

```python
#access specifier in variables
class parent:
    def __init__(self):
        self.pub_var = "Public"
        self._prot_var = "Protected"
        self.__priv_var = "Private"

    def acc_same_class(self):
        print(self.pub_var)
        print(self._prot_var)
        print(self.__priv_var)

class child_class(parent):
    def acc_sub_class(self):
        print(self.pub_var)
        print(self._prot_var)
        try:
           print(self.__priv_var)
        except AttributeError:
            print("Private variable cannot be used in the sub child class")   

#to work on the other files instead we created
# a normal class but while creating a class we passing the object of the parent class

class other_file_class:
    def acc_other_fil(self,obj):
        print(obj.pub_var)
        print(obj._prot_var)
        print(obj._parent__priv_var) #name mangling
        try:
            print(obj.__priv_var)
        except:
            print("Other calling")
        

a=parent()
a.acc_same_class()
b=child_class()
b.acc_sub_class()
print(dir(a))
c=other_file_class()
c.acc_other_fil(a)
```

```python
#access specifier in methods
class parent:
    def pub_meth(self):
        print("Public methos")

    def __prv_meth(self):
        print("Private methos")

    def _pro_meth(self):
        print("Protected methos")

    def acc_same_class(self):
        print(self.pub_meth())
        print(self.__prv_meth())
        print(self._pro_meth())

class child_class(parent):
    def acc_sub_class(self):
        print(self.pub_meth())
        print(self._pro_meth())
        try:
           print(self.__prv_meth())
        except AttributeError:
            print("Private method cannot be used in the sub child class")   

#to work on the other files instead we created
# a normal class but while creating a class we passing the object of the parent class

class other_file_class:
    def acc_other_fil(self,obj):
        print(obj.pub_meth())
        print(obj._pro_meth())
        print(obj._parent__prv_meth()) #name mangling
        try:
            print(obj.__prv_meth())
        except:
            print("Other calling")
        

a=parent()
a.acc_same_class()
b=child_class()
b.acc_sub_class()
c=other_file_class()
c.acc_other_fil(a)
```

```python
#ENCAPSULATION

class Order:
  def __init__(self,customer_name,items,total_amount,discount):
    self.customer_name = customer_name
    self.items = items
    self.__total_amount = total_amount #private var
    self.__discount = discount #private var
    
  
  def __calculate_final(self): #private method
    return self.__total_amount - self.__discount
  
  def _admin_view(self):
    return {
      "Customer Name:" : self.customer_name,
      "Items:" : self.items,
      "Total Amount:" : self.__total_amount,
      "Discount Amount" : self.__discount,
      "Final Bill:" : self.__calculate_final()
    }
    
  def customer_view(self):
    return{
      "Customer Name:" : self.customer_name,
      "Items:" : self.items,
      "Final Bill:" : self.__calculate_final()
    }
    
    
class AdminPortal:
  def show_order(self,order):
    return order._admin_view()
    

class CustomerPortal:
  def show_order(self,order):
    return order.customer_view()    
    

order =   Order("Pradiksha",["Parotta","Grill"],500,250)
admin=AdminPortal()
cust=CustomerPortal()

print("Admin Calls")
print(admin.show_order(order))#protected allowed
    

print("Customer Calls")
print(cust.show_order(order))  #private
    
    
   #instance VS class VS statis
class MyClass:
  Original_name = "Biriyani"
  def instance_method(self):
    return f"Hi this is instance method"
  
  @classmethod
  def class_method(cls,name):
    Original_name = name
    return Original_name
    
  @staticmethod
  def stat_method(q,g):
    return q+g 

my=MyClass()
print(MyClass.class_method("Parotta"))#class itself called not called using a object
print(MyClass.stat_method(5,5))
print(my.instance_method())#normal methos calling using the object



```
