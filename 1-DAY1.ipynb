```python
#*args we give as a paramaters for function 
#doesn't need to specify the defined argument 

def add(*args):
    tot = 0
    for num in args:
        tot +=num
    return tot


print(add(11,24,45,34))


#key value pair - **kwargs 
def profile (**kwargs):
    print("User Profile Details")
    for key,value in kwargs.items():
        print(f"{key} : {value}")
profile(student="Pradiksha",Age = 25,Hobby = "Painting")
```

```python
#List Method- mutable,can add data anywhere,can be any data type
toy_list = ["Dora","Bujji","Iesha"]
print(toy_list)
toy_list.append("Benny")
print(toy_list)

toy_list.insert(2,"Kulla Nari")
print(toy_list)
toy_list.remove("Iesha")
print(toy_list)

#Pop will remove last value 
toy_list.pop()
print(toy_list)

#reverse 
toy_list.reverse()
print(toy_list)

#count
toy_list = ["Dora","Bujji","Iesha","Dora"]
print("Count:",toy_list.count("Dora"))

toy_list = ["Dora","Bujji","Iesha","Benny","Kulla Nary","Diago","Teeku","Vayasana Troll"]
#List Slicing
print("Top 2 Character",toy_list[:2])
print("Last 2 Character",toy_list[-2:])
print("Mid 2 Character",toy_list[2:5])



#List iteration
print("Dora the Explorere Characters:")
i=1
for charac in toy_list:
    print(f"Character {i} : {charac}")
    i+=1

#ENUMERATE i is index position , location name also 
for i,location in enumerate(toy_list):
    print(f"Location {i} : {location}")

```
