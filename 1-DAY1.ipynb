```python
#variable scope

#1. L -> E -> G ->B
#1.LOCAL VARIABLE - L

def orders():
    food = 'Biriyani'
    print("Your Lunch is :", food)
orders()

#if i call the food outside the fucntion it will not work as it is local variable
#like this 
#print(food) will not work

#2. Enclosive
# used in the nested function 
# here we can use the e variable inside the nested function this is enclosive variable
def cards():
    discount = 11 #E
    def checkout():
          print("Your discount amt : ", discount)

    checkout()
    
cards()


#3. Global Variable

user_id = 'Pradiksha' #G

def home():
     print("Home",user_id)
def last():
     print("Last:",user_id)

home()     
last()


#4. B- Build Variable
#only works on python file not in jupyter file
#print(__file__)


#use case payments
portal = "Gpay" #G
def notification():
     receiver = "Sangeetha" #E
     def message():
          money = 2000 #L
          #f is a formatted string
          print(f"The Money worth of {money} has been sucessfully transfered to {receiver} using {portal}")
     message()
notification()     
```
